## Data Cleaning

- Clean target variable "readmitted"
- Convert target variable "readmitted" into boolean
- Missing Values
- Values with "?"
- Convert categorical data

### Convert Target Variable

```python
print(df["readmitted"].unique())
```

```python
# convert to string
df["readmitted"] = df["readmitted"].astype(str).str.strip()
```

```python
# convert boolean
df["readmitted"] = df["readmitted"].map({
    "<30": 1,  # readmitted before 30 days 
    ">30": 0,  # readmitted after 30 days 
    "NO": 0
})

df["readmitted"]
```

### Values with '?'

```python
cols_with_question = df.columns[df.isin(['?']).any()].tolist()
print(cols_with_question)
```

Replace with nan

```python
import numpy as np

df.replace('?', np.nan, inplace=True)
```

### Missing Values

```python
print( (df.isna().sum() / len(df) ) * 100)
```

Dropping the missing rows is not an option due to the fact that includes 80% of our data. Solution is filling in missing values.

```python
print(df["max_glu_serum"].unique())
```

```python
print(df["A1Cresult"].unique())
```

```python
df.fillna("None", inplace=True)
```

```python
print( (df.isna().sum() / len(df) ) * 100)
```

### Convert Categorical Data

```python
df_cat = df.select_dtypes(include='object')
df_cat.head(5)
```

```python
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder()
df_cat_1hot = cat_encoder.fit_transform(df_cat)
df_cat_1hot.toarray()
```

## Correlations

```python
corr = df.corr(numeric_only=True)["readmitted"].sort_values(ascending=False).head(10)
print(corr)
```

```python
import matplotlib.pyplot as plt

# select numeric columns
numeric_df = df.select_dtypes(include="number")

# correlation matrix
corr = numeric_df.corr()

plt.figure(figsize=(10,8))
plt.imshow(corr)

plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")

plt.tight_layout()
plt.show()
```
